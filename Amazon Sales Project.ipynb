```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = sns.countplot(x = 'Size', data = df)
```

```python
a = sns.countplot(x = 'Size', data = df)
for boxs in a.containers:
    a.bar_label(boxs)
```

```python
a = sns.countplot(x = 'Size', data = df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
# Group BY :
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
df.groupby(['Size'], as_index=True)['Qty'].sum()
```

```python
df.groupby(['Size'], as_index = False)['Qty'].sum()
```

```python
S_Qty = df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x = 'Size', y = 'Qty', data = S_Qty)
```

```python
a1 = df['Fulfilment'].value_counts()

fig, ax = plt.subplots()
ax.pie(a1, labels = a1.index, autopct = '%1.1f%%', radius = 0.7, wedgeprops = dict(width = 0.6))
ax.set(aspect = 'equal')
plt.show()
```

```python
# SCATTER PLOT

x_data = df['Category']
y_data = df['Size']
plt.figure(figsize=(10,5))
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x = 'ship-state')
plt.xlabel('Ship-State')
plt.ylabel('Count')
plt.title('Disturbution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_State = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_State.index)], x = 'ship-state')
plt.xlabel('Ship-State')
plt.ylabel('Count')
plt.title('Disturbution of State')
plt.xticks(rotation=45)
plt.show()
```
